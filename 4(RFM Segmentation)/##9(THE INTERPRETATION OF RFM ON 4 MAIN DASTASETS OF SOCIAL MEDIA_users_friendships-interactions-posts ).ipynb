```python
import pandas as pd

#### Step 1: Import libraries and load datasets
users = pd.read_csv(r"C:\Users\Sina\Desktop\CODES\Sources\4(CHATGPT LABELED OF THE COURSE)\8(users).csv")
friendships = pd.read_csv(r"C:\Users\Sina\Desktop\CODES\Sources\4(CHATGPT LABELED OF THE COURSE)\8(friendships).csv")
interactions = pd.read_csv(r"C:\Users\Sina\Desktop\CODES\Sources\4(CHATGPT LABELED OF THE COURSE)\8(interactions).csv")
posts = pd.read_csv(r"C:\Users\Sina\Desktop\CODES\Sources\4(CHATGPT LABELED OF THE COURSE)\8(posts).csv")
## Changing the objects of dates of the datasets to Datetime
interactions['Timestamp']= pd.to_datetime(interactions['Timestamp'])
posts['Timestamp']= pd.to_datetime(posts['Timestamp'])

users
```

```python
#### Step 2: Merge interactions with posts

interactions_posts = interactions.merge(posts, on='Post_ID', how='left')
#### Step 3: Rename columns to avoid confusion
interactions_posts = interactions_posts.rename(columns={
    'Username_x': 'Interacting_Username',   # from interactions dataset
    'Username_y': 'Post_Author',            # from posts dataset
    'Timestamp_x': 'Interaction_Timestamp',
    'Timestamp_y': 'Post_Timestamp'
})
interactions_posts
```

```python
#### Step 4: Add demographic info for the interacting user
interactions_full = interactions_posts.merge(users, left_on='Interacting_Username', right_on='Username', how='left')
interactions_full
```

```python
#### Step 5 (Optional): Add demographic info for the post author too
interactions_full = interactions_full.merge(
    users,
    left_on='Post_Author',
    right_on='Username',
    how='left',
    suffixes=('_Interactor', '_Author')
)
```

```python
interactions_full
```

```python
#### Step 6 (Optional): Add friendship data
# Add a column to flag if interaction is with a friend
friendships['is_friend'] = 1

# Merge to check if interacting user is a friend of the post author
interactions_full = interactions_full.merge(
    friendships,
    left_on=['Interacting_Username', 'Post_Author'],
    right_on=['Username', 'Friend'],
    how='left'
)

# Replace NaN with 0 (means not a friend)
interactions_full['is_friend'] = interactions_full['is_friend'].fillna(0).astype(int)
```

```python
interactions_full 
```

```python
#### Step 7: Prepare Interaction_Timestamp column
interactions_full['Interaction_Timestamp'] = pd.to_datetime(interactions_full['Interaction_Timestamp'])
```

```python
#### Step 8: Define Monetary values for Interaction Types
interaction_values = {
    'like': 1,
    'comment': 2,
    'share': 3
}

# Map the Interaction_Type to numeric scores
interactions_full['Monetary_Value'] = interactions_full['Interaction_Type'].map(interaction_values)
```

```python
#### Step 9: Find the most recent interaction date in the dataset
max_date = interactions_full['Interaction_Timestamp'].max()
print("Latest interaction date in dataset:", max_date)
```

```python
#### Step 10: Calculate R, F, M values per user

"""Group by the interacting user and calculate:

    Recency: Days since last interaction (relative to max date)

    Frequency: Number of interactions

    Monetary: Sum of weighted interaction values """

rfm = interactions_full.groupby('Interacting_Username').agg(
    Recency = ('Interaction_Timestamp', lambda x: (max_date - x.max()).days),
    Frequency = ('Interaction_Type', 'count'),
    Monetary = ('Monetary_Value', 'sum')
).reset_index()
```

```python
rfm
```

```python
# Create quartile-based scores (1-4) for each RFM metric
rfm['R_Quartile'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])  # Lower Recency is better, so reversed
rfm['F_Quartile'] = pd.qcut(rfm['Frequency'], 4, labels=[1,2,3,4])
rfm['M_Quartile'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])

# Combine quartiles to create an overall RFM score
rfm['RFM_Score'] = rfm['R_Quartile'].astype(str) + rfm['F_Quartile'].astype(str) + rfm['M_Quartile'].astype(str)

rfm
```

```python
# Assuming rfm dataframe and interactions_full dataframe exist
# interactions_full contains demographic info (like Age_Interactor)

# Step 1: Convert RFM_Score to numeric for easier sorting (optional)
rfm['RFM_Score_Num'] = rfm['RFM_Score'].astype(int)

# Step 2: Define RFM groups (you can tweak thresholds as needed)
# For simplicity, let's say:
# Highest = top 25% by RFM_Score_Num
# Middle = middle 50%
# Lowest = bottom 25%

quantiles = rfm['RFM_Score_Num'].quantile([0.25, 0.75])
low_thresh = quantiles[0.25]
high_thresh = quantiles[0.75]

def rfm_group(score):
    if score <= low_thresh:
        return 'Low'
    elif score >= high_thresh:
        return 'High'
    else:
        return 'Middle'

rfm['RFM_Group'] = rfm['RFM_Score_Num'].apply(rfm_group)

# Step 3: Merge rfm with demographic info (age)
# Extract user age from interactions_full (unique users only)
user_ages = interactions_full[['Interacting_Username', 'Age_Interactor']].drop_duplicates()

rfm_age = rfm.merge(user_ages, left_on='Interacting_Username', right_on='Interacting_Username', how='left')

# Step 4: Calculate age statistics per group
age_stats = rfm_age.groupby('RFM_Group')['Age_Interactor'].agg(['mean', 'median', 'min', 'max', 'count'])

print(age_stats)
```
